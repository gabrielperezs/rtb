```python
def preprocess_data(df):
    result = df.copy()
    
    one_hot_col_names = ['ad_slot_visibility', 
                            'browser', 
                            'ad_exchange', 
                            'device', 
                            'os', 
                            'region_id'
                            # 'ad_slot_id' # whoa, this one is large
                        ] 
    
    one_hot_cols = pd.get_dummies(data[one_hot_col_names])
    
    result = pd.concat([result, one_hot_cols], axis=1)
    
    result.drop(one_hot_col_names, axis=1, inplace=True)
    
    # extract date features
    data['year'] = data['timestamp'].apply(lambda ts: ts.year)
    data['month'] = data['timestamp'].apply(lambda ts: ts.month)
    data['day'] = data['timestamp'].apply(lambda ts: ts.day)
    data['weekday'] = data['timestamp'].apply(lambda ts: ts.weekday)
    
    # ad_slot has only one value, and user agent and timestamp were parsed before
    result.drop(['user_agent', 'ad_slot', 'ad_slot_id', 'timestamp'], axis=1, inplace=True)
    
    return result

data_preproc = preprocess_data(data)
data_preproc.dtypes
```
